````python
from dotenv import load_dotenv
import os

from langchain_gigachat.chat_models import GigaChat

load_dotenv()

GIGACHAT_API_KEY = os.getenv("GIGACHAT_CREDENTIALS")
if not GIGACHAT_API_KEY:
    raise ValueError("âŒ GIGACHAT_CREDENTIALS Ğ½Ğµ Ğ½Ğ°Ğ¹Ğ´ĞµĞ½ Ğ² .env")

langchain_llm = GigaChat(
    credentials=GIGACHAT_API_KEY,
    verify_ssl_certs=False,
    model="GigaChat-Max",   
    temperature=0.4,
    max_tokens=2048,
)

print("âœ… GigaChat LangChain LLM Ğ¸Ğ½Ğ¸Ñ†Ğ¸Ğ°Ğ»Ğ¸Ğ·Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½")
````

````python
from types import SimpleNamespace
from typing import List, Dict

import autogen
from autogen import AssistantAgent, UserProxyAgent, LLMConfig


class GigaChatLangChainClient:
    """
    ĞœĞ¸Ğ½Ğ¸Ğ¼Ğ°Ğ»ÑŒĞ½Ñ‹Ğ¹ ĞºĞ»Ğ¸ĞµĞ½Ñ‚ Ğ´Ğ»Ñ AG2, ĞºĞ¾Ñ‚Ğ¾Ñ€Ñ‹Ğ¹ Ğ²Ğ½ÑƒÑ‚Ñ€Ğ¸ Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµÑ‚ langchain_llm (GigaChat).
    """

    def __init__(self, config, **kwargs):
        self.model_name = config.get("model", "giga-chat-langchain")

    def create(self, params):
        """
        AG2 Ğ¿ĞµÑ€ĞµĞ´Ğ°Ñ‘Ñ‚ ÑÑĞ´Ğ° messages Ğ¸ Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ñ‹.
        ĞœÑ‹ Ğ²Ñ‹Ğ·Ñ‹Ğ²Ğ°ĞµĞ¼ langchain_llm Ğ¸ Ğ²Ğ¾Ğ·Ğ²Ñ€Ğ°Ñ‰Ğ°ĞµĞ¼ Ğ¾Ğ±ÑŠĞµĞºÑ‚ Ñ .choices[...].message.content.
        """

        # params â€” ÑĞ»Ğ¾Ğ²Ğ°Ñ€ÑŒ, ĞºĞ¾Ñ‚Ğ¾Ñ€Ñ‹Ğ¹ AutoGen Ğ¿ĞµÑ€ĞµĞ´Ğ°Ñ‘Ñ‚ Ğ² ĞºĞ»Ğ¸ĞµĞ½Ñ‚.
        # Ğ’ Ğ½Ñ‘Ğ¼ ĞµÑÑ‚ÑŒ ĞºĞ»ÑÑ‡ "messages" â€” ÑĞ¿Ğ¸ÑĞ¾Ğº ÑĞ¾Ğ¾Ğ±Ñ‰ĞµĞ½Ğ¸Ğ¹ Ğ² ÑÑ‚Ğ¸Ğ»Ğµ OpenAI:
        messages: List[Dict[str, str]] = params["messages"]
        # Ğ¿Ñ€ĞµĞ²Ñ€Ğ°Ñ‰Ğ°ĞµĞ¼ ÑĞ¾Ğ¾Ğ±Ñ‰ĞµĞ½Ğ¸Ñ Ğ² Ğ¾Ğ´Ğ¸Ğ½ Ñ‚ĞµĞºÑÑ‚Ğ¾Ğ²Ñ‹Ğ¹ Ğ¿Ñ€Ğ¾Ğ¼Ğ¿Ñ‚, Ñ‚Ğ°Ğº ĞºĞ°Ğº Ğ½Ğ°Ñˆ langchain_llm Ğ² Ñ‚ĞµĞºÑƒÑ‰ĞµĞ¹ ĞºĞ¾Ğ½Ñ„Ğ¸Ğ³ÑƒÑ€Ğ°Ñ†Ğ¸Ğ¸ Ğ¾Ğ¶Ğ¸Ğ´Ğ°ĞµÑ‚ Ğ¾Ğ´Ğ½Ñƒ ÑÑ‚Ñ€Ğ¾ĞºÑƒ
        prompt = "\n".join(f"{m['role']}: {m['content']}" for m in messages)

        # Ğ²Ñ‹Ğ·Ñ‹Ğ²Ğ°ĞµĞ¼ Ñ€ĞµĞ°Ğ»ÑŒĞ½ÑƒÑ Ğ¼Ğ¾Ğ´ĞµĞ»ÑŒ
        text = langchain_llm.predict(prompt)

        # ÑƒĞ¿Ğ°ĞºĞ¾Ğ²Ñ‹Ğ²Ğ°ĞµĞ¼ Ñ€ĞµĞ·ÑƒĞ»ÑŒÑ‚Ğ°Ñ‚ Ğ² Ñ„Ğ¾Ñ€Ğ¼Ğ°Ñ‚, Ğ¿Ğ¾Ñ…Ğ¾Ğ¶Ğ¸Ğ¹ Ğ½Ğ° OpenAI
        response = SimpleNamespace()
        response.model = self.model_name
        response.choices = []

        choice = SimpleNamespace()
        msg = SimpleNamespace()
        msg.content = text
        msg.function_call = None
        choice.message = msg

        response.choices.append(choice)
        return response

        # Ğ¡Ñ‚Ñ€ÑƒĞºÑ‚ÑƒÑ€Ğ°:
        # response = {
        # model: "giga-chat-langchain",
        # choices: [
        #     {
        #     message: {
        #         content: "Ğ¢ĞµĞºÑÑ‚ Ğ¾Ñ‚Ğ²ĞµÑ‚Ğ° Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸",
        #         function_call: None
        #     }
        #     }
        # ]
        # }

    # Ğ­Ñ‚Ğ¾Ñ‚ Ğ¼ĞµÑ‚Ğ¾Ğ´ AutoGen Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµÑ‚, Ñ‡Ñ‚Ğ¾Ğ±Ñ‹ Ğ´Ğ¾ÑÑ‚Ğ°Ñ‚ÑŒ Ñ‚ĞµĞºÑÑ‚ Ğ¾Ñ‚Ğ²ĞµÑ‚Ğ° Ğ¸Ğ· Ğ¾Ğ±ÑŠĞµĞºÑ‚Ğ° response.
    def message_retrieval(self, response):
        return [choice.message.content for choice in response.choices]

    # Ğ­Ñ‚Ğ¾Ñ‚ Ğ¼ĞµÑ‚Ğ¾Ğ´ Ğ½ÑƒĞ¶ĞµĞ½, ĞµÑĞ»Ğ¸ Ğ½Ğ°Ğ´Ğ¾ ÑÑ‡Ğ¸Ñ‚Ğ°Ñ‚ÑŒ ÑÑ‚Ğ¾Ğ¸Ğ¼Ğ¾ÑÑ‚ÑŒ Ğ¸ ĞºĞ¾Ğ»-Ğ²Ğ¾ Ñ‚Ğ¾ĞºĞµĞ½Ğ¾Ğ². 0 Ğ§Ñ‚Ğ¾Ğ±Ñ‹ AutoGen Ğ½Ğµ Ğ¿Ğ°Ğ´Ğ°Ğ»
    def cost(self, response) -> float:
        return 0.0

    # Ğ¡Ğ»ÑƒĞ¶ĞµĞ±Ğ½Ñ‹Ğ¹ Ğ¼ĞµÑ‚Ğ¾Ğ´ Ğ´Ğ»Ñ ÑĞ±Ğ¾Ñ€Ğ° ÑÑ‚Ğ°Ñ‚Ğ¸ÑÑ‚Ğ¸ĞºĞ¸, ĞºĞ¾Ğ»-Ğ²Ğ° Ñ‚Ğ¾ĞºĞµĞ½Ğ¾Ğ² Ğ¸ Ñ‚Ğ´. Ğ¾Ñ‚Ğ´Ğ°Ñ‘Ğ¼ Ğ¿ÑƒÑÑ‚Ğ¾Ğ¹ ÑĞ»Ğ¾Ğ²Ğ°Ñ€ÑŒ - AutoGen ÑÑ‚Ğ¾Ğ³Ğ¾ Ğ´Ğ¾ÑÑ‚Ğ°Ñ‚Ğ¾Ñ‡Ğ½Ğ¾.
    @staticmethod
    def get_usage(response):
        return {}


print("âœ… GigaChatLangChainClient Ğ¾Ğ¿Ñ€ĞµĞ´ĞµĞ»Ñ‘Ğ½")
````

````python
def run_pipeline_with_manager(user_query: str, max_steps: int = 8):
    planner_reply = None
    coder_reply = None
    exec_output = None
    final_report = None

    step = 0

    while step < max_steps:
        step += 1
        print(f"\n=== Ğ¨ĞĞ“ {step} ===")

        # Ğ¡Ğ¾Ğ±Ğ¸Ñ€Ğ°ĞµĞ¼ Ñ‚ĞµĞºÑƒÑ‰ĞµĞµ ÑĞ¾ÑÑ‚Ğ¾ÑĞ½Ğ¸Ğµ Ğ² Ğ¾Ğ´Ğ¸Ğ½ Ñ‚ĞµĞºÑÑ‚ Ğ´Ğ»Ñ Ğ¼ĞµĞ½ĞµĞ´Ğ¶ĞµÑ€Ğ°
        state_text = (
            f"Ğ—Ğ°Ğ¿Ñ€Ğ¾Ñ Ğ¿Ğ¾Ğ»ÑŒĞ·Ğ¾Ğ²Ğ°Ñ‚ĞµĞ»Ñ:\n{user_query}\n\n"
            f"Ğ¢Ğ— Ğ¾Ñ‚ planner (Ğ¼Ğ¾Ğ¶ĞµÑ‚ Ğ±Ñ‹Ñ‚ÑŒ Ğ¿ÑƒÑÑ‚Ñ‹Ğ¼):\n{planner_reply}\n\n"
            f"ĞŸĞ¾ÑĞ»ĞµĞ´Ğ½Ğ¸Ğ¹ ÑĞ³ĞµĞ½ĞµÑ€Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ½Ñ‹Ğ¹ ĞºĞ¾Ğ´ (Ğ¼Ğ¾Ğ¶ĞµÑ‚ Ğ±Ñ‹Ñ‚ÑŒ Ğ¿ÑƒÑÑ‚Ñ‹Ğ¼):\n{coder_reply}\n\n"
            f"ĞŸĞ¾ÑĞ»ĞµĞ´Ğ½Ğ¸Ğ¹ Ğ²Ñ‹Ğ²Ğ¾Ğ´ Ğ¸ÑĞ¿Ğ¾Ğ»Ğ½ĞµĞ½Ğ¸Ñ exec_output (Ğ¼Ğ¾Ğ¶ĞµÑ‚ Ğ±Ñ‹Ñ‚ÑŒ Ğ¿ÑƒÑÑ‚Ñ‹Ğ¼):\n{exec_output}\n\n"
            f"ĞŸĞ¾ÑĞ»ĞµĞ´Ğ½Ğ¸Ğ¹ Ğ¸Ñ‚Ğ¾Ğ³Ğ¾Ğ²Ñ‹Ğ¹ Ğ¾Ñ‚Ñ‡Ñ‘Ñ‚ (Ğ¼Ğ¾Ğ¶ĞµÑ‚ Ğ±Ñ‹Ñ‚ÑŒ Ğ¿ÑƒÑÑ‚Ñ‹Ğ¼):\n{final_report}\n\n"
            "Ğ ĞµÑˆĞ¸, Ñ‡Ñ‚Ğ¾ Ğ´ĞµĞ»Ğ°Ñ‚ÑŒ Ğ´Ğ°Ğ»ÑŒÑˆĞµ, Ğ¸ Ğ²ĞµÑ€Ğ½Ğ¸ JSON ÑĞ¾Ğ³Ğ»Ğ°ÑĞ½Ğ¾ Ğ¸Ğ½ÑÑ‚Ñ€ÑƒĞºÑ†Ğ¸Ğ¸."
        )

        manager_reply = manager.generate_reply(
            messages=[{"role": "user", "content": state_text}]
        )
        print("\nğŸ§­ ĞÑ‚Ğ²ĞµÑ‚ manager:\n", manager_reply)

        # ĞŸĞ°Ñ€ÑĞ¸Ğ¼ JSON Ğ¾Ñ‚ manager
        try:
            import json
            manager_json = json.loads(manager_reply)
        except Exception as e:
            print("âš ï¸ ĞĞµ ÑƒĞ´Ğ°Ğ»Ğ¾ÑÑŒ Ñ€Ğ°ÑĞ¿Ğ°Ñ€ÑĞ¸Ñ‚ÑŒ JSON Ğ¾Ñ‚ manager, Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞ°ĞµĞ¼:", e)
            break

        action = manager_json.get("action")
        fb_planner = manager_json.get("feedback_for_planner") or ""
        fb_coder = manager_json.get("feedback_for_coder") or ""
        print("â¡ï¸ Ğ”ĞµĞ¹ÑÑ‚Ğ²Ğ¸Ğµ:", action)

        if action == "PLAN":
            # Ğ’Ñ‹Ğ·Ğ¾Ğ² planner Ñ ÑƒÑ‡Ñ‘Ñ‚Ğ¾Ğ¼ Ñ„Ğ¸Ğ´Ğ±ĞµĞºĞ°
            planner_prompt = user_query
            if fb_planner:
                planner_prompt += f"\n\nĞ”Ğ¾Ğ¿Ğ¾Ğ»Ğ½Ğ¸Ñ‚ĞµĞ»ÑŒĞ½Ñ‹Ğµ Ğ·Ğ°Ğ¼ĞµÑ‡Ğ°Ğ½Ğ¸Ñ Ğº Ñ„Ğ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ğ¸:\n{fb_planner}\n"
            planner_reply = planner.generate_reply(
                messages=[{"role": "user", "content": planner_prompt}]
            )
            print("\nğŸ§  ĞĞ¾Ğ²Ñ‹Ğ¹ Ğ¾Ñ‚Ğ²ĞµÑ‚ planner:\n", planner_reply)

        elif action == "CODE":
            if not planner_reply:
                print("âš ï¸ ĞĞµÑ‚ Ğ¢Ğ— Ğ¾Ñ‚ planner, ÑĞ½Ğ°Ñ‡Ğ°Ğ»Ğ° Ğ½ÑƒĞ¶ĞµĞ½ PLAN.")
                continue
            coder_prompt = (
                "ĞĞ¸Ğ¶Ğµ Ñ„Ğ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ¾Ğ²Ğ°Ğ½Ğ½Ğ¾Ğµ Ğ¢Ğ— Ğ¾Ñ‚ Ğ°Ğ½Ğ°Ğ»Ğ¸Ñ‚Ğ¸ĞºĞ°.\n"
                "ĞĞ°Ğ¿Ğ¸ÑˆĞ¸ ĞºĞ¾Ğ´ ÑĞ¾Ğ³Ğ»Ğ°ÑĞ½Ğ¾ ÑĞ²Ğ¾Ğ¸Ğ¼ Ğ¸Ğ½ÑÑ‚Ñ€ÑƒĞºÑ†Ğ¸ÑĞ¼.\n\n"
                "=== Ğ¢Ğ— Ğ°Ğ½Ğ°Ğ»Ğ¸Ñ‚Ğ¸ĞºĞ° ===\n"
                f"{planner_reply}\n"
                "=== ĞšĞ¾Ğ½ĞµÑ† Ğ¢Ğ— ===\n"
            )
            if fb_coder:
                coder_prompt += f"\nĞ”Ğ¾Ğ¿Ğ¾Ğ»Ğ½Ğ¸Ñ‚ĞµĞ»ÑŒĞ½Ñ‹Ğµ Ğ·Ğ°Ğ¼ĞµÑ‡Ğ°Ğ½Ğ¸Ñ Ğº ĞºĞ¾Ğ´Ñƒ:\n{fb_coder}\n"

            coder_reply = coder.generate_reply(
                messages=[{"role": "user", "content": coder_prompt}]
            )
            print("\nğŸ’» ĞĞ¾Ğ²Ñ‹Ğ¹ ĞºĞ¾Ğ´ coder:\n", coder_reply)

        elif action == "EXECUTE":
            if not coder_reply:
                print("âš ï¸ ĞĞµÑ‚ ĞºĞ¾Ğ´Ğ° Ğ´Ğ»Ñ Ğ¸ÑĞ¿Ğ¾Ğ»Ğ½ĞµĞ½Ğ¸Ñ, ÑĞ½Ğ°Ñ‡Ğ°Ğ»Ğ° Ğ½ÑƒĞ¶ĞµĞ½ CODE.")
                continue
            exec_output = executor.generate_reply(
                messages=[{"role": "user", "content": coder_reply}]
            )
            print("\nâ–¶ï¸ Ğ’Ñ‹Ğ²Ğ¾Ğ´ Ğ¸ÑĞ¿Ğ¾Ğ»Ğ½ĞµĞ½Ğ¸Ñ exec_output:\n", exec_output)

        elif action == "MAKE_REPORT":
            # Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·ÑƒĞµĞ¼ ÑƒĞ¶Ğµ ÑÑƒÑ‰ĞµÑÑ‚Ğ²ÑƒÑÑ‰Ğ¸Ğ¹ result_agent, Ğ½Ğ¾ Ñ‚ĞµĞ¿ĞµÑ€ÑŒ Ğ¾Ğ½ Ğ¿Ñ€Ğ¾ÑÑ‚Ğ¾ Ğ´ĞµĞ»Ğ°ĞµÑ‚ Ğ¾Ñ‚Ñ‡Ñ‘Ñ‚,
            # Ğ° Ğ½Ğµ Ñ€ĞµÑˆĞ°ĞµÑ‚, Ñ‡Ğ¸Ğ½Ğ¸Ñ‚ÑŒ Ğ»Ğ¸ ĞºĞ¾Ğ´
            if not exec_output:
                print("âš ï¸ ĞĞµÑ‚ exec_output, Ğ¿Ğ¾ĞºĞ° Ñ€Ğ°Ğ½Ğ¾ Ğ´ĞµĞ»Ğ°Ñ‚ÑŒ Ğ¾Ñ‚Ñ‡Ñ‘Ñ‚.")
                continue

            result_prompt = (
                "Ğ¡Ñ„Ğ¾Ñ€Ğ¼Ğ¸Ñ€ÑƒĞ¹ Ğ°Ğ½Ğ°Ğ»Ğ¸Ñ‚Ğ¸Ñ‡ĞµÑĞºĞ¸Ğ¹ Ğ¾Ñ‚Ñ‡Ñ‘Ñ‚ Ğ´Ğ»Ñ Ğ¿Ğ¾Ğ»ÑŒĞ·Ğ¾Ğ²Ğ°Ñ‚ĞµĞ»Ñ Ğ½Ğ° Ğ¾ÑĞ½Ğ¾Ğ²Ğµ Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ… Ğ½Ğ¸Ğ¶Ğµ.\n\n"
                f"Ğ—Ğ°Ğ¿Ñ€Ğ¾Ñ Ğ¿Ğ¾Ğ»ÑŒĞ·Ğ¾Ğ²Ğ°Ñ‚ĞµĞ»Ñ:\n{user_query}\n\n"
                f"Ğ¢Ğ— Ğ¾Ñ‚ planner:\n{planner_reply}\n\n"
                f"ĞšĞ¾Ğ´, ĞºĞ¾Ñ‚Ğ¾Ñ€Ñ‹Ğ¹ Ğ²Ñ‹Ğ¿Ğ¾Ğ»Ğ½ÑĞ»ÑÑ:\n{coder_reply}\n\n"
                f"Ğ’Ñ‹Ğ²Ğ¾Ğ´ Ğ¸ÑĞ¿Ğ¾Ğ»Ğ½ĞµĞ½Ğ¸Ñ (stdout/Ğ¾ÑˆĞ¸Ğ±ĞºĞ¸):\n{exec_output}\n\n"
                "Ğ•ÑĞ»Ğ¸ ĞµÑÑ‚ÑŒ Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¸ Ğ¾ÑˆĞ¸Ğ±ĞºĞ¸ Ğ¸ÑĞ¿Ğ¾Ğ»Ğ½ĞµĞ½Ğ¸Ñ â€” ÑƒĞ¿Ğ¾Ğ¼ÑĞ½Ğ¸ ÑÑ‚Ğ¾ Ğ² Ğ¾Ñ‚Ñ‡Ñ‘Ñ‚Ğµ."
            )

            final_report = result_agent.generate_reply(
                messages=[{"role": "user", "content": result_prompt}]
            )
            print("\nğŸ“Š Ğ˜Ñ‚Ğ¾Ğ³Ğ¾Ğ²Ñ‹Ğ¹ Ğ¾Ñ‚Ñ‡Ñ‘Ñ‚ (ÑÑ‹Ñ€Ğ¾Ğ¹ Ñ‚ĞµĞºÑÑ‚):\n", final_report)
            # ĞœĞ¾Ğ¶Ğ½Ğ¾ ÑÑ€Ğ°Ğ·Ñƒ Ğ¾Ñ‚Ğ¾Ğ±Ñ€Ğ°Ğ¶Ğ°Ñ‚ÑŒ ĞºÑ€Ğ°ÑĞ¸Ğ²Ğ¾:
            display(Markdown(final_report))

        elif action == "FINISH":
            print("â¹ Manager Ñ€ĞµÑˆĞ¸Ğ» Ğ¾ÑÑ‚Ğ°Ğ½Ğ¾Ğ²Ğ¸Ñ‚ÑŒÑÑ.")
            break

        else:
            print("âš ï¸ ĞĞµĞ¸Ğ·Ğ²ĞµÑÑ‚Ğ½Ğ¾Ğµ Ğ´ĞµĞ¹ÑÑ‚Ğ²Ğ¸Ğµ, Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞ°ĞµĞ¼.")
            break

    return {
        "planner_reply": planner_reply,
        "coder_reply": coder_reply,
        "exec_output": exec_output,
        "final_report": final_report,
    }
````
