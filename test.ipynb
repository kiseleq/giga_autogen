```python
from dotenv import load_dotenv
import os

# –ó–∞–≥—Ä—É–∂–∞–µ–º .env
load_dotenv()

GIGACHAT_API_KEY = os.getenv("GIGACHAT_CREDENTIALS")
if not GIGACHAT_API_KEY:
    raise ValueError("‚ùå GIGACHAT_CREDENTIALS –Ω–µ –Ω–∞–π–¥–µ–Ω –≤ .env")

# –ò–º–ø–æ—Ä—Ç –º–æ–¥–µ–ª–∏
from langchain_gigachat.chat_models import GigaChat

# –°–æ–∑–¥–∞—ë–º LLM
langchain_llm = GigaChat(
    credentials=GIGACHAT_API_KEY,
    verify_ssl_certs=False,
    model="GigaChat",
    temperature=0.4,
    max_tokens=256,
)

```

```python
question = "–ü—Ä–∏–≤–µ—Ç! –ö–∞–∫ –¥–µ–ª–∞?"

# –ü–æ–ª—É—á–∞–µ–º –æ—Ç–≤–µ—Ç
response = langchain_llm.predict(question)

# ---------- –í–´–í–û–î ----------
print(response)
```
